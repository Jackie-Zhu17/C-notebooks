```python
###Apriori算法定义###
def createC1(dataSet):   ##生成单个的一项集
    '''元素个数为1的项集'''
    C1 = []
    for transaction in dataSet:
        for item in transaction:
            if [item] not in C1:
                C1.append([item])
    C1.sort()
    return list(map(frozenset, C1))

def scanD(D, Ck, minSupport):   ##D为set（dataset），扫描数据集，找出满足minsup的组成Ck项集
    '''Ck候选项集'''
    ssCnt = {}
    for tid in D:
        for can in Ck:
            if can.issubset(tid):
                ssCnt[can] = ssCnt.get(can, 0) + 1        ###统计各项的次数
    numItems = float(len(D))
    retList = []
    supportData = {}
    for key in ssCnt:
        support = ssCnt[key] / numItems
        if support >= minSupport:
            retList.insert(0, key)
        
        supportData[key] = support
    return retList, supportData

def aprioriGen(Lk, k):
    '''该函数通过频繁项集列表Lk和项集个数k生成候选项集Ck+1。

注意其生成的过程中，首选对每个项集按元素排序，然后每次比较两个项集，只有在前k-1项相同时才将这两项合并。这样做是因为函数并非要两两合并各个集合，那样生成的集合并非都是k+1项的。在限制项数为k+1的前提下，只有在前k-1项相同、最后一项不相同的情况下合并才为所需要的新候选项集。

由于Python中使用下标0表示第一个元素，因此代码中的[:k-2]的实际作用为取列表的前k-1个元素。'''
    retList = []
    lenLk = len(Lk)
    for i in range(lenLk):
        for j in range(i + 1, lenLk):
            # 前k-2项相同时，将两个集合合并
            L1 = list(Lk[i])[:k-2]; L2 = list(Lk[j])[:k-2]
            L1.sort(); L2.sort()
            if L1 == L2:
                retList.append(Lk[i] | Lk[j])
    return retList

def apriori(dataSet, minSupport=0.5):
    '''该函数为Apriori算法的主函数，按照前述伪代码的逻辑执行。Ck表示项数为k的候选项集，最初的C1通过createC1()函数生成。Lk表示项数为k的频繁项集，supK为其支持度，Lk和supK由scanD()函数通过Ck计算而来。

函数返回的L和supportData为所有的频繁项集及其支持度，因此在每次迭代中都要将所求得的Lk和supK添加到L和supportData中。'''
    C1 = createC1(dataSet)
    D = list(map(set, dataSet))
    L1, supportData = scanD(D, C1, minSupport)
    L = [L1]
    k = 2
    while (len(L[k-2]) > 0):
        Ck = aprioriGen(L[k-2], k)
        Lk, supK = scanD(D, Ck, minSupport)
        supportData.update(supK)
        L.append(Lk)
        k += 1
    return L, supportData


C1 = createC1(tags)
D = list(map(set, tags))
L1, suppDat = scanD(D, C1, 0.295)
print(L1)
L, suppData = apriori(tags, minSupport=0.09)  ###最小支持度###
print(L)
L_ = pd.DataFrame({'tags':L},index = [u'1项集',u'2项集',u'3项集'])
L_   
###L_为频繁项集###
```

```python
###评论中的关联规则###

def generateRules(L, supportData, minConf=0.705):    ###置信度###
    bigRuleList = []
    for i in range(1, len(L)):
        for freqSet in L[i]:
            H1 = [frozenset([item]) for item in freqSet]
            if (i > 1):
                # 三个及以上元素的集合
                rulesFromConseq(freqSet, H1, supportData, bigRuleList, minConf)
            else:
                # 两个元素的集合
                calcConf(freqSet, H1, supportData, bigRuleList, minConf)
    return bigRuleList

def calcConf(freqSet, H, supportData, brl, minConf=0.705):
    ''' 对候选规则集进行评估 '''
    prunedH = []
    for conseq in H:
        conf = supportData[freqSet] / supportData[freqSet - conseq]
        if conf >= minConf:
            a = pd.DataFrame({u'前项':freqSet - conseq,u'后项':conseq,u'置信度':conf})
            print(a)
            brl.append((freqSet - conseq, conseq, conf))
            prunedH.append(conseq)
    return prunedH

def rulesFromConseq(freqSet, H, supportData, brl, minConf=0.705):
    ''' 生成候选规则集 '''
    m = len(H[0])
    if (len(freqSet) > (m + 1)):
        Hmpl = aprioriGen(H, m + 1)
        Hmpl = calcConf(freqSet, Hmpl, supportData, brl, minConf)
        if (len(Hmpl) > 1):
            rulesFromConseq(freqSet, Hmpl, supportData, brl, minConf)
            
rules = generateRules(L, suppData, minConf=0.3)
```

```python
###寻找项集与刷单数###

def find(Lk,data,support=0.295):
    std , bustd = 0,0
    target_result = []
    for m in Lk:
        for n in range(len(data)):
            if m.issubset(data[n]):
                if Y[n] == 0:
                    bustd += 1
                else :
                    std+=1
        if std/float(bustd + std) < support:
            continue
        target_result.append([m,bustd,std])
        std,bustd = 0,0
            
    return target_result
```

```python
from sklearn import feature_extraction  
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer  

tags_new = [' '.join(i) for i in tags]
vectorizer=CountVectorizer()#该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频  
transformer=TfidfTransformer()#该类会统计每个词语的tf-idf权值  
tfidf=transformer.fit_transform(vectorizer.fit_transform(tags_new))#第一个fit_transform是计算tf-idf，第二个fit_transform是将文本转为词频矩阵  
word=vectorizer.get_feature_names()#获取词袋模型中的所有词语  
weight=tfidf.toarray()#将tf-idf矩阵抽取出来，元素a[i][j]表示j词在i类文本中的tf-idf权重  
```

```python
print('Start Kmeans:')
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=30)
s = kmeans.fit(weight)
print(s)
```

```python
def train(X,vectorizer,true_k=10,showLable = False):   ###showLable控制是否打印出具体标签###
    '''kmeans训练'''   
    km = KMeans(n_clusters=true_k, init='k-means++', max_iter=300, n_init=1,verbose=False)
    km.fit(X)    
    if showLable:
        print('Top terms per cluster:')
        order_centroids = km.cluster_centers_.argsort()[:, ::-1]
        terms = vectorizer.get_feature_names()  ###根据TF-IDF提取特征词
        print ('stop words:',vectorizer.get_stop_words())
        for i in range(true_k):
            print("Cluster %d:" % i, end='')
            for ind in order_centroids[i, :10]:  ###每个簇前m个词条###
                print(' %s' % terms[ind], end='')
            print()
    result = list(km.predict(X))
    print ('Cluster distribution:')   ###每个簇多少样本###
    print (dict([(i, result.count(i)) for i in result]))
    return -km.score(X)

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_df=0.5, max_features=500, min_df=2,use_idf=True)
w = vectorizer.fit_transform(tags_new)
score = train(w,vectorizer,true_k=10,showLable=True)/w.shape[0]
```
